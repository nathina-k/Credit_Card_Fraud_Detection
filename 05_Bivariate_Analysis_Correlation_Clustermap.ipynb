## <a id="d_ID"> 4. Bivariate Analysis</a>

### <a id="d3_ID">4.3 Spearman Correlation Matrix</a>

The Spearman correlation matrix tells us quantatively which numeric features are correlated with each other.

```python
# Spearman correlation matrix
creditcard.corr(method="spearman")
```

### <a id="d4_ID">4.4 Spearman Correlation Heatmap</a>

A heatmap describes corrlations between numeric features by colour and its intensity.

```python
# Spearman correlation heatmap
plt.figure(figsize=(20,20))
sns.heatmap(creditcard.corr(method = 'spearman'), cmap = "YlGnBu", annot=True)
plt.show()
```

### <a id="d5_ID">4.5 Clustermap</a>

A clustermap describes corrlations between numeric features by colour and dendrogram.

```python
# Clustermap
sns.clustermap(creditcard.corr(method="spearman"), cmap='YlGnBu', annot=True, figsize=(15, 15))
plt.show()
```

```python
# Import scipy library
import scipy.stats
```

```python
# Highest positive correlation from heatmap and clustermap
print(scipy.stats.spearmanr(creditcard['V21'], creditcard['V22']))
```

```python
# Highest negative correlation from heatmap and clustermap
print(scipy.stats.spearmanr(creditcard['V2'], creditcard['Amount']))
```

The highest positive correlation in our data is 0.68 between V21 and V22 (moderate positive correlation). The highest negative correlation is -0.50 between V2 and Amount (moderate negative correlation). Both are highly statistically significant with a p-value of 0.0.
