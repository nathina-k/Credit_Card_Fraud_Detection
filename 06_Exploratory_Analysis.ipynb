### <a id="e1_ID">5.1 K-means Clustering</a>

K-means clustering is an unsupervised machine learning algorithm used to classify data into clusters. Data points within a cluster are similar to each other and data points from separate clusters are different from each other.

```python
# Create dataframe with subset dataset of numeric features (excludes Class attribute)
kmeansdata = pd.DataFrame(creditcard, columns = ["Time", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9", "V10", "V11", "V12", "V13", "V14", "V15", "V16", "V17", "V18", "V19", "V20", "V21", "V22", "V23", "V24", "V25", "V26", "V27", "V28", "Amount"])
kmeansdata
```

```python
# Import KMeans library
from sklearn.cluster import KMeans
```

```python
# Instantiate KMeans object with n_clusters=2
kmeansModel2 = KMeans(n_clusters=2, random_state=31)
```

```python
# Fit dataset to KMeans object
kmeansModel2.fit(kmeansdata)
```

```python
# Show cluster centers
kmeansModel2.cluster_centers_
```

```python
# Show cluster labels
kmeansModel2.labels_
```

```python
# Instantiate KMeans object with n_clusters=4
kmeansModel4 = KMeans(n_clusters=4, random_state=32)
```

```python
# Fit dataset to KMeans object
kmeansModel4.fit(kmeansdata)
```

```python
# Show cluster centers
kmeansModel4.cluster_centers_
```

```python
# Show cluster labels
kmeansModel4.labels_
```

```python
# Scatterplots of K-means with 2 clusters, 4 clusters, and by Class
plt.figure(figsize=(20,5))
plt.subplot(1, 3, 1)
plt.scatter(kmeansdata.loc[:,'V1'], kmeansdata.loc[:,'V2'], c=kmeansModel2.labels_, 
            cmap='rainbow', alpha=0.2)
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('2 Clusters Identified with K Means')

plt.subplot(1, 3, 2)
plt.scatter(kmeansdata.loc[:,'V1'], kmeansdata.loc[:,'V2'], c=kmeansModel4.labels_, 
            cmap='rainbow', alpha=0.2)
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('4 Clusters Identified with K Means')

plt.subplot(1, 3, 3)
plt.scatter(creditcard.loc[:,'V1'], creditcard.loc[:,'V2'], c=creditcard.loc[:,'Class'], 
            label=creditcard.loc[:,'Class'].unique(), cmap='rainbow', alpha=0.2)
plt.xlabel('V1')
plt.ylabel('V2')
plt.title('Clusters Identified By Class')

plt.show()
```

From the figures above, we can see that two clusters (red for genuine cases and purple for fraudlent cases) exactly identifies the data points categorized by Class, genuine transactions (purple) vs fraudulent transactions (red). Four clusters tell us how each the genuine and fraudulent transactions can be further broken down into two groups (turquoise/red for genuine cases and light green/purple for fraudulent cases).
